```python
# Firefox on your system.

from selenium.webdriver.firefox.service import Service
from webdriver_manager.firefox import GeckoDriverManager
# Set up Firefox WebDriver
def fetch_page_content(url):
    # Set up Firefox WebDriver
    driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()))
    
    # Open the webpage
    driver.get('https://www.spinny.com/used-cars-in-jaipur/s/')
    
    # Wait for the page to load (adjust sleep time as needed)
    time.sleep(10)
    
    # Get the page source after JavaScript has rendered the content
    page_source = driver.page_source
    
    # Close the browser
    driver.quit()
    
    return page_source

# Parse the page source with BeautifulSoup
def parse_page_content(html_content):
    return BeautifulSoup(html_content, 'lxml')
```

#### Combine the code accordingly.

```python
def extract_car_details(soup):
    company = soup.find_all('div', class_='CarListingCardV2__carListingCarContainer')
    
    names = []
    years = []
    prices = []
    km_driven = []
    states = []
    transmissions = []
    fuel_types = []
    emi_per_month = []

    for car in company:
        # Extract name and year
        h3_tag = car.find('h3', class_='ListingBrandModelDetail__makeModelInfo')
        if h3_tag:
            make_span = h3_tag.find('span', class_='ListingBrandModelDetail__make')
            if make_span:
                full_text = make_span.text.strip()
                parts = full_text.split()
                if len(parts) > 1:
                    year = parts[0]
                    company_name = ' '.join(parts[1:])
                    names.append(company_name)
                    years.append(year)
                else:
                    names.append("N/A")
                    years.append("N/A")
            else:
                names.append("N/A")
                years.append("N/A")
        else:
            names.append("N/A")
            years.append("N/A")

        # Extract price
        ul_tag = car.find('ul')
        if ul_tag:
            price_span = ul_tag.find('span', class_='ListingPricingDetail__priceWithRupeeSymbol')
            if price_span:
                prices.append(price_span.text.strip())
            else:
                prices.append("N/A")
        else:
            prices.append("N/A")

        # Extract EMI per month
        emi_span = car.find('li', class_='ListingPricingDetail__emi')
        if emi_span:
            emi_per_month.append(emi_span.text.strip()[5:-3])
        else:
            emi_per_month.append("N/A")

        # Extract fuel type, transmission, state, and km driven
        detail_ul = car.find('ul', class_='CarListingCardDetail__more')
        if detail_ul:
            details = detail_ul.find_all('li')
            fuel_type = "N/A"
            transmission = "N/A"
            state = "N/A"
            km = "N/A"
            for detail in details:
                text = detail.text.strip().lower()
                if 'petrol' in text or 'diesel' in text or 'electric' in text:
                    fuel_type = text
                elif 'automatic' in text or 'manual' in text:
                    transmission = text
                elif re.match(r'^[A-Z]{2}\d{1,2}', text.upper()):  # Match state codes like DL, HR, RJ, etc.
                    state = text.upper()[:2]  # Extract only the first two characters (state code)
                elif 'km' in text:  # Check for km driven
                    km = text
            fuel_types.append(fuel_type)
            transmissions.append(transmission)
            states.append(state)
            km_driven.append(km)
        else:
            fuel_types.append("N/A")
            transmissions.append("N/A")
            states.append("N/A")
            km_driven.append("N/A")

    return names, years, prices, km_driven, states, transmissions, fuel_types, emi_per_month
```

# Step-5. Function to save data to a CSV file

```python
# Function to save data to a CSV file
def save_to_csv(data, filename='car_data.csv'):
    df = pd.DataFrame(data)
    df.to_csv(filename, index=False)

# Main function to run the script
def main():
    url = 'https://www.spinny.com/used-cars-in-jaipur/s/'
    
    # Fetch page content using Selenium
    web_page = fetch_page_content(url)
    soup = parse_page_content(web_page)
    
    # Extract car details
    names, years, prices, km_driven, states, transmissions, fuel_types, emi_per_month = extract_car_details(soup)
    
    # Create a dictionary to store the data
    data = {
        'Name': names,
        'Year': years,
        'Price': prices,
        'Kilometers Driven': km_driven,
        'State': states,
        'Transmission': transmissions,
        'Fuel Type': fuel_types,
        'EMI per Month': emi_per_month
    }
    
    # Save the data to a CSV file
    save_to_csv(data)
    print("Data saved to car_data.csv")

if __name__ == "__main__":
    main()
```
